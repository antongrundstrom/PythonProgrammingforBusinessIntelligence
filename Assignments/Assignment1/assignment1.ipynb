## Task 1: A list of Dictonaries

In this task, I filter out invalid transactions, calculate the total value for each valid transaction, group revenue by region, and identify the region with the highest total revenue.

```python
transactions = [
    {"id": 1, "product": "Laptop", "price": 1200, "quantity": 2, "region": "EU"},
    {"id": 2, "product": "Phone", "price": 800, "quantity": -1, "region": "US"},
    {"id": 3, "product": "Tablet", "price": 0, "quantity": 5, "region": "EU"},
    {"id": 4, "product": "Monitor", "price": 350, "quantity": 3, "region": "US"},
    {"id": 5, "product": "Laptop", "price": 1200, "quantity": 1, "region": "EU"},
    {"id": 6, "product": "Keyboard", "price": 75, "quantity": 10, "region": "APAC"},
]
```

```python
# a) We start by filtering out invalid transactions
valid_transactions = []

for transaction in transactions:
    if transaction["price"] > 0 and transaction["quantity"] >= 0:
        valid_transactions.append(transaction.copy())            # I made copies to not override the original "Transaction"

print("Valid transactions:")
for transaction in valid_transactions:
    print(transaction)

# b) Then for each remaining transaction we compute a new field (total = price * quantity)
transactions_with_total = []

for transaction in valid_transactions:
    new_transaction = transaction.copy()
    new_transaction["total"] = new_transaction["price"] * new_transaction["quantity"]
    transactions_with_total.append(new_transaction)

print("\nValid transactions with total:")               #"\n adds an empty row to make the final result easier to understand
for transaction in transactions_with_total:
    print(transaction)

# c) We then group by region and compute the total revenue per region
revenue_by_region = {}

for transaction in transactions_with_total:
    region = transaction["region"]
    total = transaction["total"]

    if region not in revenue_by_region:
        revenue_by_region[region] = 0

    revenue_by_region[region] += total

print("\nRevenue by region:")
print(revenue_by_region)

# d) Lastly, here we print the region with the highest total revenue using an f-string
highest_region = max(revenue_by_region, key=revenue_by_region.get)
highest_value = revenue_by_region[highest_region]

print(f"\nThe region with the highest total revenue is {highest_region} with {highest_value}")
```

## Task 2: Creating a function

In this task, I create a function called `summarise_sales` that takes a list of daily sales values and returns a dictionary with summary statistics.

```python
sales = [1200, 980, 1540, 870, 2010, 1650, 1300, 1100, 950, 1780, 2200, 1430, 1600, 1900]
```

```python
def summarise_sales(sales_list):
    total = sum(sales_list)
    average = round(total / len(sales_list), 2)
    best_day = sales_list.index(max(sales_list)) + 1     #We add + 1 because indexing in Python starts at 0
    worst_day = sales_list.index(min(sales_list)) + 1
    
    above_average_count = 0
    for sale in sales_list:
        if sale > average:
            above_average_count += 1
    
    return {
        "total": total,
        "average": average,
        "best_day": best_day,
        "worst_day": worst_day,
        "above_average_count": above_average_count
    }

result = summarise_sales(sales)

print(summarise_sales(sales))
```

## Task 3: Loops

In this task, I use `for` loops and `if` statements to filter students, print status messages, count attendance below a threshold, and find the student with the highest grade both with and without `max()`.

```python
students = [
    {"name": "Anna", "grade": 85, "attendance": 0.92},
    {"name": "Erik", "grade": 58, "attendance": 0.75},
    {"name": "Mia", "grade": 91, "attendance": 0.98},
    {"name": "Leo", "grade": 44, "attendance": 0.60},
    {"name": "Sara", "grade": 72, "attendance": 0.88},
    {"name": "Tom", "grade": 67, "attendance": 0.70},
    {"name": "Nina", "grade": 95, "attendance": 0.99},
    {"name": "Olli", "grade": 50, "attendance": 0.65},
]
```

```python
# a) We create a list called passing of students with grade 50 or above
passing = []

for student in students:
    if student["grade"] >= 50:
        passing.append(student["name"])

print("Passing students:")
print(passing)

# b) Here we group the grades into status messages
print("\nStudent results:")
for student in students:
    name = student["name"]
    grade = student["grade"]
    
    if grade >= 80:
        print(f"{name}: Excellent")
    elif 50 <= grade <= 79:
        print(f"{name}: Pass")
    else:
        print(f"{name}: Fail")

# c) Next we count students with attendance below 0.80 and print the count
low_attendance_count = 0

for student in students:
    if student["attendance"] < 0.80:
        low_attendance_count += 1

print(f"\nNumber of students with attendance below 0.80: {low_attendance_count}")

# d) Find the student with the highest grade using a loop
top_student_loop = students[0]

for student in students:
    if student["grade"] > top_student_loop["grade"]:
        top_student_loop = student

print(f"\nTop student using a loop: {top_student_loop['name']} with grade {top_student_loop['grade']}")

# e) Find the student with the highest grade using max()
top_student_max = max(students, key=lambda student: student["grade"])

print(f"Top student using max(): {top_student_max['name']} with grade {top_student_max['grade']}")
```

## Task 4: Class and methods

Here I create a class called `Portfolio` to represent an investment portfolio. The class can store assets, calculate the total portfolio value, find the most valuable asset, and print a readable summary.

```python
class Portfolio:
    def __init__(self, owner):
        self.owner = owner
        self.holdings = {}  # ticker -> {shares, price_per_share}

    # a) Add a new asset or update an existing one
    def add_asset(self, ticker, shares, price_per_share):
        self.holdings[ticker] = {
            "shares": shares,
            "price_per_share": price_per_share
        }

    # b) Return the total market value of the portfolio
    def total_value(self):
        total = 0
        for ticker in self.holdings:
            shares = self.holdings[ticker]["shares"]
            price = self.holdings[ticker]["price_per_share"]
            total += shares * price
        return total

    # c) Return the ticker with the highest total value
    def most_valuable(self):
        highest_ticker = None
        highest_value = 0

        for ticker in self.holdings:
            shares = self.holdings[ticker]["shares"]
            price = self.holdings[ticker]["price_per_share"]
            asset_value = shares * price

            if asset_value > highest_value:
                highest_value = asset_value
                highest_ticker = ticker

        return highest_ticker

    # d) Return a readable summary of the portfolio
    def __str__(self):
        return f"Portfolio owner: {self.owner}, Number of assets: {len(self.holdings)}, Total value: {self.total_value()}"
```

## Task 5: Importing a module

In task 5, I use the `datetime` module to convert date strings into `datetime.date` objects and then calculate the number of days between the earliest and latest order dates.

```python
from datetime import datetime

orders = [
    ("2025-01-15", "Laptop"), ("2025-01-22", "Phone"),
    ("2025-02-03", "Laptop"), ("2025-02-14", "Tablet"),
    ("2025-03-01", "Phone"), ("2025-03-10", "Phone"),
    ("2025-03-22", "Laptop"), ("2025-04-05", "Monitor"),]

# Parse each date string into a datetime.date object
parsed_dates = []

for order in orders:
    date_string = order[0]
    date_object = datetime.strptime(date_string, "%Y-%m-%d").date()
    parsed_dates.append(date_object)

print("Parsed dates:")
for date in parsed_dates:
    print(date)

# Find earliest and latest date
earliest_order = min(parsed_dates)
latest_order = max(parsed_dates)

# Compute the number of days between the orders
days_between = (latest_order - earliest_order).days

print(f"\nEarliest order date: {earliest_order}")
print(f"Latest order date: {latest_order}")
print(f"Number of days between earliest and latest order: {days_between}")
```

## Task 6: Array Operations and Broadcasting

In this task, I use NumPy to calculate row sums, column means, apply broadcasting to adjust the 2024 revenue values, and use boolean indexing to find revenues above 250.

```python
# Task 6: Array Operations and Broadcasting

import numpy as np

# Rows: years (2022, 2023, 2024), Columns: quarters (Q1, Q2, Q3, Q4)
revenue = np.array(
    [[210, 230, 190, 250],
    [220, 260, 200, 270],
    [250, 290, 230, 310]])

# a) Total annual revenue for each year (row sums)
annual_revenue = np.sum(revenue, axis=1)
print("a) Total annual revenue for each year:")
print(annual_revenue)

# b) Mean quarterly revenue across all years (column means)
mean_quarterly_revenue = np.mean(revenue, axis=0)
print("\nb) Mean quarterly revenue across all years:")
print(mean_quarterly_revenue)

# c) Apply growth adjustment to the 2024 row using broadcasting
growth_adjustment = np.array([1.05, 1.03, 1.08, 1.02])
adjusted_2024 = revenue[2] * growth_adjustment

print("\nc) Adjusted 2024 revenue values:")
print(adjusted_2024)

# d) Find all quarterly revenue values that exceed 250
high_revenue = revenue[revenue > 250]
print("\nd) Quarterly revenue values that exceed 250:")
print(high_revenue)
```

## Task 7: Pandas

In task 7, I create a pandas DataFrame, add a calculated column, sort the data, filter rows using `.loc`, rename a column, and select specific rows and columns using `.iloc`.

```python
import pandas as pd

data = {
    "employee": ["Anna", "Erik", "Mia", "Leo", "Sara", "Tom"],
    "department": ["Finance", "Marketing", "Finance", "IT", "Marketing", "IT"],
    "salary": [4500, 4000, 4800, 5200, 3900, 5100],
    "years_exp": [5, 3, 7, 10, 2, 8],
    "performance_score": [4.2, 3.8, 4.5, 4.0, 3.5, 4.9]}
```

```python
# a) Add a new column salary_per_year_exp
df = pd.DataFrame(data)
df["salary_per_year_exp"] = (df["salary"] / df["years_exp"]).round().astype(int)

print("a) DataFrame with salary_per_year_exp:")
df[["employee", "salary", "years_exp", "salary_per_year_exp"]]
```

```python
# b) Sort by performance_score in descending order and print top 3 employees
top_3 = df.sort_values(by="performance_score", ascending=False).head(3)

print("b) Top 3 employees by performance score:")
top_3[["employee", "performance_score"]]
```

```python
# c) Select Finance employees with performance score above 4.0 using .loc
finance_high_perf = df.loc[
    (df["department"] == "Finance") & (df["performance_score"] > 4.0),
    ["employee", "department", "performance_score"]
]

print("c) Finance employees with performance score above 4.0:")
finance_high_perf
```

```python
# d) Rename years_exp to experience_years
df = df.rename(columns={"years_exp": "experience_years"})

print("d) DataFrame after renaming years_exp to experience_years:")
df
```

```python
# e) Using .iloc, select salary and performance score for the first four employees
print("e) Salary and performance score for the first four employees:")
df.iloc[:4][["salary", "performance_score"]]
```

## Task 8: More Pandas

Lastly, in task 8, I generate daily website traffic data, create a pandas DataFrame with the date as the index, and compute the number of conversions and daily revenue.

```python
import numpy as np
import pandas as pd

rng = np.random.default_rng(seed=123)
days = pd.date_range(start="2025-01-06", periods=28, freq="D")
visitors = rng.integers(200, 600, size=28)
conversion_rate = rng.uniform(0.02, 0.08, size=28).round(4)
avg_order_value = rng.normal(50, 10, size=28).round(2)

df = pd.DataFrame({
    "Date": days,
    "Visitors": visitors,
    "Conversion_Rate": conversion_rate,
    "Avg_Order_Value": avg_order_value})

df = df.set_index("Date")

print("a) DataFrame with date as index:")
df.head()
```

```python
df["Conversions"] = (df["Visitors"] * df["Conversion_Rate"]).round().astype(int)
df["Daily_Revenue"] = df["Conversions"] * df["Avg_Order_Value"]

print("b) DataFrame with conversions and daily_revenue:")
df.head()
```
